```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
import numpy as np
import seaborn as sns
import os
```

```python
NoDeadend = pd.read_csv("NoDeadend.csv", sep = ';')
Deadend = pd.read_csv("Deadend.csv", sep = ';')
```

```python
NoDeadend.drop(NoDeadend.tail(1).index,inplace=True)
NoDeadend
```

```python
x = NoDeadend['number of free cells at the start']
x
```

```python
y = NoDeadend['solution time']
y
```

```python
z = NoDeadend['memory used (MB)']
z
```

```python
X_Y_Spline = make_interp_spline(x, y)
X_ND = np.linspace(x.min(), x.max(), 500)
Y_ND = X_Y_Spline(X_ND)

plt.plot(X_ND, Y_ND)
plt.title("Number of free cells at the start VS Solution time \n (Without Deadend)")
plt.xlabel("Number of free cells at the start")
plt.ylabel("Solution time")
plt.tight_layout()
plt.savefig("NoDeadend_Graph_time.png")
plt.show()
```

```python
X_Z_Spline = make_interp_spline(x, z)
X_ND_M = np.linspace(x.min(), x.max(), 500)
Z_ND_M = X_Z_Spline(X_ND_M)

plt.plot(X_ND_M, Z_ND_M)
plt.title("Number of free cells at the start VS Memory used (MB)\n (Without Deadend)")
plt.xlabel("Number of free cells at the start")
plt.ylabel("Memory used (MB)")
plt.tight_layout()
plt.savefig("NoDeadend_Graph_memory.png")
plt.show()
```
